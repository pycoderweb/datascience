```python
df = pd.read_csv("/content/cities_air_quality_water_pollution.18-10-2021.csv")
df.head()
```

```python
df.columns
```

```python
df.columns = ['City', 'Region', 'Country', 'AirQuality','WaterPollution']
df.head()
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"',''))
df.head()
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"',''))
df.head()
```

```python
df['Region'] = df['Region'].map(lambda x: x.strip())
df.head()
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip())
df.head()
df[df.Country=='Germany']
```

```python
df.head()
```

```python
df1 = pd.pivot_table(df,index=['Country','Region'])
df1
```

```python
df1['AirPollution'] = 100 - df1['AirQuality']
df1['WaterQueality'] = 100 - df1['WaterPollution']
df1.head()
```

```python
df2 = pd.DataFrame(df1.values,index=df1.index,columns=[['Air','Air','Water','Water'],['AirQuality', 'WaterPollution', 'AirPollution', 'WaterQueality']])
df2
```

```python
df2.columns.names=['Object','Qua_Poll']
df2
```
