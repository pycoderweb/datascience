### 5-VAZIFA. AQSh 2020-yilda jami qancha elektr energiyasi ishlab chiqqan?

```python
# JAVOBNI SHUYERGA YOZING
def jami_energiya(d):
  j=0
  for i in np.array(dfus.columns):
    if i not in ['davlat','kod','yil']:
      j+=dfus.at[d,f'{i}']
  return j

jami_energiya(2020)
```

### 9-VAZIFA. 2019-yilda har bir davlat jami qancha elektr energiya ishlab chiqarganini DF ko'rinishida chiqaring.
*ishora. `Series` obyektini `DF` ga o'tkazing


```python
# # JAVOBNI SHUYERGA YOZING
myser = df.loc[df['yil'] == 2019,['davlat', 'gaz', 'komir', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']]
myser['max'] = myser[['gaz', 'komir', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].sum(axis=1)
df1 = myser[['davlat', 'max']]
df11 = pd.DataFrame(df1['max'].to_numpy(), index=df1['davlat'], columns=['max'])
df11
```
